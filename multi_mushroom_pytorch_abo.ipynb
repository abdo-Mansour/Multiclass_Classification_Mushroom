```python
# Read in labels from csv for train dataset
image_files = []
labels = []
images_path = '/kaggle/input/mushroom-multiclass-classification/dataset/dataset/'
labels_path = '/kaggle/input/mushroom-multiclass-classification/train.csv'

with open(labels_path, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        img_name = row[0]
        label = row[1]
        img_path = os.path.join(images_path, img_name+'.jpg')
        if os.path.exists(img_path):
            image_files.append(img_path)
            labels.append(label)

```
